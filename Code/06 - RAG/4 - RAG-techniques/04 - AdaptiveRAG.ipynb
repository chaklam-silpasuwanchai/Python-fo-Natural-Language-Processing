# [Adaptive-RAG: Learning to Adapt Retrieval-Augmented Large Language Models through Question Complexity](https://arxiv.org/pdf/2403.14403)

It addresses the limitations of traditional Retrieval-Augmented Generation (RAG) by introducing a classifier that routes queries into three different execution paths based on complexity:
- Single-Step (Simple Queries): For straightforward questions like "When is Michael Phelps' birthday?", the model performs a single retrieval step to find the answer.
- Multi-Step (Complex Queries): For multi-hop questions like "What is the currency of Billy Giles' birthplace?", the model performs iterative retrieval ($k$ times) to gather intermediate facts before answering.
- No Retrieval (Straightforward Queries): For common knowledge (e.g., "Paris is the capital of what?"), the LLM answers directly without searching documents to save time and resources.
